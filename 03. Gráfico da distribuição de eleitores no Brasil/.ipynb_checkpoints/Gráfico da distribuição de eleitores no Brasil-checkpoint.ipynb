# IMPORTAÇÃO DOS DADOS

## Fonte dos dados
> Foram utilizados dados extraídos diretamente dos órgãos oficiais responsáveis.

`perfil_eleitorado_2022.zip`
> * https://dadosabertos.tse.jus.br/dataset/eleitorado-2022/resource/8a6ff7bd-5a22-4d1a-9353-cce06f9f5e35
> * última atualização: 15 de julho de 2022
> * *dataset* com as informações do eleitorado brasileiro, separados por município.

`bcim_2016_21_11_2018.gpkg`
> * https://www.ibge.gov.br/geociencias/cartas-e-mapas/bases-cartograficas-continuas/
> * última atualização: 12 de novembro de 2022, 08:55 (UTC-03:00)
> * *informações geoespaciais* com todos os pontos de localizações dos municípios do Brasil, extraídos do site do IBGE.


```python
# Carrega a base de dados de abstenção do 1ª turno das eleições 2022.
df = pd.read_csv('raw\perfil_eleitorado_2022.zip', sep=';', encoding='latin1')

# Carrega o arquivo geoespacial disponibilizado pelo governo brasileiro com os pontos cardeais dos municípios
municipios = gpd.read_file('https://geoftp.ibge.gov.br/cartas_e_mapas/bases_cartograficas_continuas/bcim/versao2016/geopackage/bcim_2016_21_11_2018.gpkg', layer = 'LIM_Municipio_A')    # Não vamos utilizar a base de municipios nessa análise
estados = gpd.read_file('https://geoftp.ibge.gov.br/cartas_e_mapas/bases_cartograficas_continuas/bcim/versao2016/geopackage/bcim_2016_21_11_2018.gpkg', layer = 'LIM_Unidade_Federacao_A')
```

# ESTRUTURA DOS DADOS


## Correções e alterações

### Correções na base de municípios

```python
# Separando os códigos de UF e Municipio em colunas separadas
municipios['CD_MUNICIPIO'] = municipios['geocodigo'].str[2:]
municipios['CD_UF'] = municipios['geocodigo'].str[:2]

# Normalizamos os nomes para minúsculos para cruzarmos as bases
municipios['nome'] = municipios['nome'].str.lower() \
                             .str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
# Checamos se existe algum nulo, pois precisamos de todos preenchidos
municipios[municipios['CD_MUNICIPIO'].isnull() == True]
```

```python
# Encontramos apenas Amapá, então checamos se é algum caso de duplicado
# Infelizmente não existe informação preenchida, então teremos que desconsiderá-lo.
municipios[municipios['nome'] != 'amapa']
municipios = municipios[municipios['nome'] != 'amapa']
```

```python
municipios.info()
```

```python
municipios['CD_MUNICIPIO'] = municipios['CD_MUNICIPIO'].astype(int)
```

```python
estados.sample(3)
```

```python
# Fazemos um dataframe novo só com os estados e os códigos, retirados da base de estados
# Precisaremos para cruzar com a base de municipio, para evitar duplicar na hora do merge 
codigos_estados = pd.DataFrame(estados.groupby('sigla')['geocodigo'].max().reset_index())
codigos_estados.head(5)
```

```python
# Fazemos o merge entre a base de eleitores e de estados.
municipios_novos = pd.merge(municipios, codigos_estados, how="inner", left_on="CD_UF", right_on="geocodigo")

municipios_novos.head()
```

```python
municipios_novos.info()
```

### Correções na base de eleitores

```python
df.sample(3)
```

```python
df['NM_MUNICIPIO'] = df['NM_MUNICIPIO'].str.lower() \
                     .str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
linhas_originais = df.shape[0]
print(f"Temos um total de {df.shape} linhas e colunas")
```

```python
# Fazemos o merge entre a base de eleitores e de estados.
df_cruzado = pd.merge(df, municipios_novos, left_on = ['NM_MUNICIPIO','SG_UF'], right_on=['nome','sigla'], how = 'left')

# Conferimos novamente que a base continua com o mesmo tamanho
print(f">>>> Temos um total de {df_cruzado.shape[0]} linhas, com uma diferença de {df_cruzado.shape[0] - linhas_originais} para a base original")
```

```python
df_cruzado.sample(5)
```

# VISÃO GERAL

```python
# Agrupamos o dataframe por cidade somando a quantidade de eleitores (QT_ELEITORES_PERFIL),
# pois identificamos que o arquivo está separado por faixa etária, genero, grau de escolaridade, etc.

# Dessa forma teremos uma linha por UF com o total de eleitores
# Vamos apresentar os dados por milhão, para facilitar a visualização

df_municipio = df_cruzado.groupby(['NM_MUNICIPIO','SG_UF'], as_index=False)['QT_ELEITORES_PERFIL'].sum().sort_values(by='QT_ELEITORES_PERFIL',  ascending=False)
df_municipio['QT_ELEITORES_PERFIL'] = df_municipio['QT_ELEITORES_PERFIL']
df_municipio['QT_ELEITORES_PERFIL_PERC'] = df_municipio['QT_ELEITORES_PERFIL'] / df_municipio['QT_ELEITORES_PERFIL'].sum()
df_municipio['QT_ELEITORES_PERFIL_PERC_100'] = (df_municipio['QT_ELEITORES_PERFIL_PERC'] * 100).round(1)
df_municipio['QT_ELEITORES_PERFIL_PERC_100_GRAFICO'] = df_municipio['QT_ELEITORES_PERFIL_PERC_100'].apply(str) + '%'

df_municipio
```

```python
# Fazemos um merge entre o df que criamos com a soma dos eleitores entre os estados, 
# e a base do IBGE, para conseguir os pontos geoespaciais para plotar o gráfico

df_municipio = pd.merge(df_municipio, municipios_novos, left_on=['NM_MUNICIPIO','SG_UF'], right_on=['nome','sigla'], how="left")
#df_municipio = df_uf.dropna()
```

```python
df_municipio.shape
```

```python
# Checando as médias para poder montar o gráfico
df_municipio.describe().T
```

```python
# Top 15 cidades com mais eleitores
df_municipio.head(30)
```

```python
# Aqui nós separamos a base de dados e criamos o objeto do Geopandas
populacao_brasil = df_municipio[['NM_MUNICIPIO','QT_ELEITORES_PERFIL','QT_ELEITORES_PERFIL_PERC','geometry']]
gdf_populacao = gpd.GeoDataFrame(populacao_brasil, geometry='geometry')
```

```python
# Montamos a figura e plotamos o mapa com o matplotlib
fig, ax = plt.subplots(figsize=(30, 25), 
                       subplot_kw=dict(aspect='equal'))
plt.gcf().text(0.15, .89, "ELEITORADO 2022 - Onde estão os eleitores no Brasil?", fontsize = 40, color='Black' ,ha='left', va='center', weight="bold")
plt.gcf().text(0.15, .87, "Distribuição dos eleitores brasileiros em quantidade por município", fontsize = 19, color='Black' ,ha='left', va='center')

gdf_populacao.plot(column='QT_ELEITORES_PERFIL', 
                   scheme='natural_breaks', 
                   k=20, 
                   cmap='BuPu', 
                   edgecolor='white',
                   linewidth=.5,
                   legend=True, 
                   legend_kwds={'loc': 'lower left'},
                   ax=ax,
                  )

plt.axis('off')
plt.show()
```

```python
# Calcular o total de eleitores
total_eleitores = df['QT_ELEITORES_PERFIL'].sum()
print(f">>>> Temos um total de {total_eleitores} eleitores.")
```

> Somando todas as linhas temos um total de **156.454.011** eleitores no Brasil, batendo com o número divulgado pelo TSE.


```python
# Setando os 2 eixos para o gráfico
eixo_x = df_municipio['NM_MUNICIPIO'].str.title().head(10)
eixo_y = df_municipio['QT_ELEITORES_PERFIL'].head(10)

# Montando o setup e plotando o gráfico
cores = ["#eac435","#f4e21a","#ffff00","#dee92c","#afd62b","#87c141","#5fad56","#2f9f62","#189868","#00916e"]

fig = plt.figure(figsize=(24,12))
plot = sns.barplot(x=eixo_x,y=eixo_y, palette=cores)
plt.gcf().text(.5, .93, "ELEITORADO - 2022", fontsize = 30, color='Black' ,ha='center', va='center', weight='bold')
plt.title("Top 10 cidades em números de eleitores", fontsize = 15, ha='center', va='center')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel(" ", size = 14)
plt.ylabel("Eleitores (em milhões)", size = 14)

plot.legend(bbox_to_anchor=(1, 1), ncol=1)
plot.set_frame_on(False)
plot.yaxis.set_major_formatter(ticker.EngFormatter())

for i in plot.patches:
    plot.annotate("{:.1f}M".format(i.get_height()/1000000),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center', va='top', fontsize='20',
                  color='black', xytext=(0,-10),
                  textcoords='offset points')          
plt.show()
```
